```python
from sklearn.metrics import roc_auc_score
from keras.callbacks import Callback

train_aucs = {}
val_aucs = {}
global_counter = 1

class roc_callback(Callback):
    def __init__(self,training_data,validation_data):
        self.x = training_data[0]
        self.y = training_data[1]
        self.x_val = validation_data[0]
        self.y_val = validation_data[1]

    def on_train_begin(self, logs={}):
        return

    def on_train_end(self, logs={}):
        return

    def on_epoch_begin(self, epoch, logs={}):
        return

    def on_epoch_end(self, epoch, logs={}):
        y_pred = self.model.predict(self.x)
        roc = roc_auc_score(self.y, y_pred)
        y_pred_val = self.model.predict(self.x_val)
        roc_val = roc_auc_score(self.y_val, y_pred_val)
        
        if train_aucs.get(global_counter) is None:
            train_aucs[global_counter] = []
            train_aucs[global_counter].append(roc)
        else:
            train_aucs[global_counter].append(roc)
       
        if val_aucs.get(global_counter) is None:
            val_aucs[global_counter] = []
            val_aucs[global_counter].append(roc_val)
        else:
            val_aucs[global_counter].append(roc_val)
            
        print('\rroc-auc: %s - roc-auc_val: %s' % (str(round(roc,4)),str(round(roc_val,4))),end=100*' '+'\n')
        return

    def on_batch_begin(self, batch, logs={}):
        return

    def on_batch_end(self, batch, logs={}):
        return
```

```python
def pad(train_data, val_data, MAX_LENGTH):
  tokenizer = Tokenizer()
  texts = train_data.values.tolist()
  tokenizer.fit_on_texts(texts)
  train_seq = tokenizer.texts_to_sequences(texts)
  
  val_seq = tokenizer.texts_to_sequences(val_data.values.tolist())
  word_index = tokenizer.word_index
  print('Found %s unique tokens.' % len(word_index))
  
  X_train_pad = pad_sequences(train_seq, maxlen = MAX_LENGTH)
  X_val_pad = pad_sequences(val_seq, maxlen = MAX_LENGTH)
  
  return X_train_pad, X_val_pad, word_index;

def pad_test(test_data, MAX_LENGTH):
  tokenizer = Tokenizer()
  texts = test_data.values.tolist()
  tokenizer.fit_on_texts(texts)
  test_seq = tokenizer.texts_to_sequences(texts)
  word_index = tokenizer.word_index
  print('Found %s unique tokens.' % len(word_index))
  
  X_test_pad = pad_sequences(test_seq, maxlen= MAX_LENGTH)
  
  return X_test_pad;

def create_word2vec_model(total_lines, EMBEDDING_DIM):
  
    model = gensim.models.Word2Vec(sentences = total_lines,
                              size = EMBEDDING_DIM,
                              #negative = 5,
                              window = 5,
                              workers = 6,
                              min_count = 1)
    
    words = list(model.wv.vocab)
    print('Vocab size: ', len(words))
    return model;

def get_embeddings_index(word2vec_model, word2vec_path):
    # Save word2vec model
    word2vec_model.wv.save_word2vec_format(word2vec_path, binary = False)
    
    embeddings_index = {}
    # Load word2vec model to get embeddings index
    with open(word2vec_path, encoding = 'utf-8') as f:
        for line in f:
            values = line.split()
            word = values[0]
            coeffs = np.asarray(values[1:])
            embeddings_index[word] = coeffs
    
    return embeddings_index;
    
def get_embedding_matrix(EMBEDDING_DIM, train_word_index, embeddings_index):
    num_words = len(train_word_index) + 1
    embedding_matrix = np.zeros((num_words, EMBEDDING_DIM))
        
    for word, i in train_word_index.items():
        if i >= num_words:
            continue
        embedding_vector = embeddings_index.get(word)
        if embedding_vector is not None:
            embedding_matrix[i] = embedding_vector
    
    return num_words, embedding_matrix;
        
def create_model(num_words, EMBEDDING_DIM, embedding_matrix, MAX_LENGTH):
  print("Building model...")
  model = Sequential()
  embedding_layer = Embedding(num_words,
                             EMBEDDING_DIM,
                             embeddings_initializer = Constant(embedding_matrix),
                             input_length = MAX_LENGTH,
                             trainable = False)

  model.add(embedding_layer)
  model.add(GRU(units = 32, dropout = 0.2, recurrent_dropout = 0.2))
  model.add(Dense(1, activation = 'sigmoid'))
  
#   adam = optimizers.Adam(lr=0.0001)
  model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
  print(model.summary())
    
  return model;

def train_model(model, X_train_pad, y_train, X_val_pad, y_val, checkpoint_file_path):
    # Get class weights and early stopping callback
    class_weights = class_weight.compute_class_weight('balanced',
                                                np.unique(y_train),
                                                y_train)
    
    class_weight_dict = dict(enumerate(class_weights))
    print(class_weight_dict)
    
    all_callbacks = [EarlyStopping(monitor = 'val_loss', patience = 5),
            ModelCheckpoint(filepath = checkpoint_file_path, 
                            monitor = 'val_loss', 
                            save_best_only = True)]
#     roc_callback(training_data = (X_train_pad, y_train), validation_data = (X_val_pad, y_val))
    print("Training model...")
    result = model.fit(X_train_pad, 
              y_train, 
              batch_size = 128,
              epochs = 20,
              validation_data = (X_val_pad, y_val), 
              verbose = 2, 
              callbacks = all_callbacks,
              class_weight = class_weight_dict, 
              shuffle = True)
    
    return result;
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report


MAX_LENGTH = max([len(s) for s in X.values.tolist()])
# print(MAX_LENGTH)

EMBEDDING_DIM = 128

# 5-stratified shuffle split, so classes are balanced
skf = StratifiedShuffleSplit(n_splits = 5, test_size = 0.4)
num_folds = skf.get_n_splits(X)

loss_scores = []
train_accs = []
val_accs = []
test_accs = []
test_aucs = []
# test_precisions = []
# test_recalls = []


for index, (train_index, val_index) in enumerate(skf.split(X, y)):
  counter = str(index + 1)
  print("Training on fold " + counter + "/" + str(num_folds) + "...")

  word2vec_path = '/Users/admin/Documents/Queens_Masters_Courses/Deep_Learning/course_project/final_models/w2v_models/base_w2v_model_{}.txt'.format(counter)
#   print(word2vec_path)

  checkpoint_file_path = '/Users/admin/Documents/Queens_Masters_Courses/Deep_Learning/course_project/final_models/checkpoint_models/base_model_checkpoint_{}.h5'.format(counter)
#   print(checkpoint_file_path)
    
  model_file_path = '/Users/admin/Documents/Queens_Masters_Courses/Deep_Learning/course_project/final_models/base_models/base_model_{}.h5'.format(counter)
#   print(model_file_path)
    
  X_train, X_val_temp = X[train_index], X[val_index]
  y_train, y_val_temp = y[train_index], y[val_index]
  X_val, X_test, y_val, y_test = train_test_split(X_val_temp, y_val_temp, test_size=.5)
#   print(len(X_train))
#   print(len(X_val))
#   print(len(X_test))
#   print(X_train)
  
  X_train_pad, X_val_pad, train_word_index = pad(X_train, X_val, MAX_LENGTH)
  X_test_pad = pad_test(X_test, MAX_LENGTH)
  
  frames_temp = [X_train, X_val]
  data_temp = pd.concat(frames_temp)
  data_temp = data_temp.reset_index(drop = True)
  total_lines = data_temp.values.tolist()

  word2vec_model = create_word2vec_model(total_lines, EMBEDDING_DIM)
 
  embeddings_index = get_embeddings_index(word2vec_model, word2vec_path)

  num_words, embedding_matrix = get_embedding_matrix(EMBEDDING_DIM, 
                                                     train_word_index, 
                                                     embeddings_index)

#   print('Shape of X_train_pad tensor:', X_train_pad.shape)
#   print('Shape of y_train tensor:', y_train.shape)

#   print('Shape of X_val_pad tensor:', X_val_pad.shape)
#   print('Shape of y_val tensor:', y_val.shape)
  
#   print('Shape of X_test_pad tensor:', X_test_pad.shape)
#   print('Shape of y_test tensor:', y_test.shape)
#   print('====================')
  model = None
  history = None
    
  model = create_model(num_words, EMBEDDING_DIM, embedding_matrix, MAX_LENGTH)
  
  history = train_model(model, X_train_pad, y_train, X_val_pad, y_val, checkpoint_file_path)
  
  training_accuracy_history = history.history['acc']
  val_accuracy_history = history.history['val_acc']
  print("Last training accuracy:" + str(training_accuracy_history[-1]) + ", validation accuracy: " + str(val_accuracy_history[-1]))
  train_accs.append(training_accuracy_history[-1])
  val_accs.append(val_accuracy_history[-1])

  # Loss score and accuracy
  test_score, test_acc = model.evaluate(X_test_pad, y_test, batch_size = 128)
  print('Test score:', test_score)
  print('Test accuracy:', test_acc)
  print("Accuracy: {0:.2%}".format(test_acc))

  loss_scores.append(test_score)
  test_accs.append(test_acc)
    
  # Test set prediction with trained model
  y_pred = model.predict(x = X_test_pad, batch_size = 128)
#   print(classification_report(y_test, y_pred))
#   # Test Precision
#   test_precision = precision_score(y_test, y_pred.round(), normalize = False)
#   print("Model's precision on test set:", test_precision)
    
#   test_precisions.append(test_precision)

#   # Test Recall
#   test_recall = recall_score(y_test, y_pred.round(), normalize = False)
#   print("Model's recall on test set:", test_recall)
    
#   test_recalls.append(test_recall)
    
  # Test AUC
  test_auc = roc_auc_score(y_test, y_pred)
  print("Model's AUC on test set:", test_auc)

  test_aucs.append(test_auc)
  
  # Save model
  model.save(model_file_path)

#   global_counter += 1
    
print('============+ Results +==============')
print("Loss scores:")
print(loss_scores)
print("Training accuracies:")
print(train_accs)
print("Validation accuracies:")
print(val_accs)
print("Testing accuracies:")
print(test_accs)
print("Test AUCs:")
print(test_aucs)
# print("Train AUCs:")
# print(train_aucs)
# print("Validation AUCs:")
# print(val_aucs)
# print("Test Precisions:")
# print(test_precisions)
# print("Test Recalls:")
# print(test_recalls)

```
